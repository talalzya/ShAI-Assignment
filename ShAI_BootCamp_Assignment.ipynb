## Data Cleaning

```python
#check if we have any null values
data.isnull().sum()
```

```python
#drop the columns that does not have any useful values 
data2 = data.drop(['Notes','Status'],axis=1)
data2
```

```python
#drop NaN
data3 = data2.dropna()
data3
```

### Basic Data Visulization

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.histplot(data=data3, x="TotalSalary", kde=True, color="skyblue")

plt.title("The Distribution of Salaries")

plt.show()
```

### Grouped Analysis

```python
grouped_data = data3.groupby('JobTitle')['TotalSalary'].agg(['mean','std' ,'min', 'max' ])
grouped_data
```

grouped_data.plot(kind='bar', xlabel='JobTitle', ylabel='Total Salary', title='Average Salary Across Groups')

### Simple Correlation Analysis

```python
correlation_coef= data3['TotalSalary'].corr(data3['OvertimePay'])
print(f"Correlation_Coeficient:{correlation_coef}")
```

```python
plt.scatter(data3['TotalSalary'],data3['OvertimePay'])
plt.xlabel('TotalSalary')
plt.ylabel('OvertimePay')
```

```python
sns.heatmap(data3.corr(), annot=True, cmap='plasma')
```
